```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
from plotly import tools
import matplotlib.pyplot as plt#visualization


import plotly.offline as py#visualization
py.init_notebook_mode(connected=True)#visualization
import plotly.graph_objs as go#visualization
import plotly.tools as tls#visualization
import plotly.figure_factory as ff#visualization
```

```python
# Import Data set
df = pd.read_csv('Bank.csv')
df.head()
```

```python
# Data Cleaning
df.shape

col = df.columns.tolist()

# Replace empty space with NA
for i in col: 
    df[i] = df[i].replace(" ", np.nan)

# Calculating percentage of Missing Data
for i in col:
    null = df[i].isnull().sum()/len(df)
    print(i, null)    

```

```python
# Separate Categorical / Numerical variables
y = ['deposit']
cat_cols = df.nunique()[df.nunique() < 13].keys().tolist()
cat_cols = [x for x in cat_cols if x not in y]
num_cols = [x for x in df.columns if x not in cat_cols + y]
```

```python
cat_cols
```

```python
depositY = df[df["deposit"]== 'yes']
depositN = df[df["deposit"]== 'no']
```

```python
# Customer Attrition in Data

# It shows that we have a balance dependent variable

# labels 
label = df['deposit'].value_counts().keys().tolist()
# Values
val = df['deposit'].value_counts().values.tolist()

trace = go.Pie(labels = label,
              values = val,
              marker = dict(colors = ['royalblue', 'lime'],
                           line = dict(color = 'white', width = 1.3)
                           ),
              rotation = 90,
              hoverinfo = "label+value+text",
              hole = 0.5
              )
layout = go.Layout(dict(title = 'Customer Attrition in data',
                       plot_bgcolor = 'rgb(243,243,243)',
                       paper_bgcolor = 'rgb(243,243,243)',
                       )
                  )
data = [trace]
fig = go.Figure(data = data, layout = layout)
py.iplot(fig)
```

```python
def plot_pie(column) :
    
    trace1 = go.Pie(values  = depositY[column].value_counts().values.tolist(),
                    labels  = depositY[column].value_counts().keys().tolist(),
                    hoverinfo = "label+percent+name",
                    domain  = dict(x = [0,.48]),
                    name    = "Deposit (Yes) Customers",
                    marker  = dict(line = dict(width = 2,
                                               color = "rgb(243,243,243)")
                                  ),
                    hole    = .6
                   )
    trace2 = go.Pie(values  = depositN[column].value_counts().values.tolist(),
                    labels  = depositN[column].value_counts().keys().tolist(),
                    hoverinfo = "label+percent+name",
                    marker  = dict(line = dict(width = 2,
                                               color = "rgb(243,243,243)")
                                  ),
                    domain  = dict(x = [.52,1]),
                    hole    = .6,
                    name    = "Deposit (No) customers" 
                   )
    layout = go.Layout(dict(title = column + " distribution in customer attrition ",
                        plot_bgcolor  = "rgb(243,243,243)",
                        paper_bgcolor = "rgb(243,243,243)",
                        annotations = [dict(text = "Deposit Yes customers",
                                                font = dict(size = 13),
                                                showarrow = False,
                                                x = .15, y = .5),
                                           dict(text = "Deposit No customers",
                                                font = dict(size = 13),
                                                showarrow = False,
                                                x = .88,y = .5
                                               )
                                          ]
                           )
                      )
    data = [trace1,trace2]
    fig  = go.Figure(data = data,layout = layout)
    py.iplot(fig)
    

def histogram(column):
    trace1 = go.Histogram(x  = depositY[column],
                          histnorm= "percent",
                          name = "Deposit (Yes) Customers",
                          marker = dict(line = dict(width = .5,
                                                    color = "black"
                                                    )
                                        ),
                         opacity = .9 
                         ) 
    
    trace2 = go.Histogram(x  = depositN[column],
                          histnorm = "percent",
                          name = "Deposit (No) customers",
                          marker = dict(line = dict(width = .5,
                                              color = "black"
                                             )
                                 ),
                          opacity = .9
                         )
    data = [trace1,trace2]
    layout = go.Layout(dict(title =column + " distribution in customer attrition ",
                            plot_bgcolor  = "rgb(243,243,243)",
                            paper_bgcolor = "rgb(243,243,243)",
                            xaxis = dict(gridcolor = 'rgb(255, 255, 255)',
                                             title = column,
                                             zerolinewidth=1,
                                             ticklen=5,
                                             gridwidth=2
                                            ),
                            yaxis = dict(gridcolor = 'rgb(255, 255, 255)',
                                             title = "percent",
                                             zerolinewidth=1,
                                             ticklen=5,
                                             gridwidth=2
                                            ),
                           )
                      )
    fig  = go.Figure(data=data,layout=layout)
    
    py.iplot(fig)
    

```

```python
# Categorical Variables Pie Charts
for i in cat_cols :
    plot_pie(i)
    
# Numerical Variables Histogram
for i in num_cols:
    histogram(i)
```

### Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# Binary Variables 
binary = df.nunique()[df.nunique() == 2].keys().tolist()
# Categorical with > 2 levels
mul = [x for x in cat_cols if x not in binary]

```

```python
# Encoding Binary Variable
label= LabelEncoder()
for i in binary:
    df[i] = label.fit_transform(df[i])
```

```python
# Create Dummy variables
df = pd.get_dummies(data = df, columns=mul)
```

```python
standard = StandardScaler()
scaled = standard.fit_transform(df[num_cols])
scaled = pd.DataFrame(scaled, columns = num_cols)
```

```python
df = df.drop(columns = num_cols, axis = 1)
df = df.merge(scaled, left_index = True, right_index = True, how = 'left')
```

### Correlation Matrix

```python
# Correlation 
cor = df.corr()
# Labels
lab = cor.columns.tolist()
# convert to array
cor_array= np.array(cor)

# Plotting
trace = go.Heatmap(z = cor_array,
                  x= lab,
                  y = lab,
                  colorscale = "Viridis",
                  colorbar = dict(title = "Pearson Correlation Coefficient",
                                 titleside = "right"
                                 ),
                  )
layout = go.Layout(dict(title = "Correlation Matrix",
                       autosize = False,
                       height = 720, width = 800,
                       margin = dict(r = 9, l = 218,
                                    t = 25, b = 210,),
                       yaxis = dict(tickfont = dict(size = 9)),
                       xaxis = dict(tickfont = dict(size = 9))
                       )
                  )
data = [trace]
fig = go.Figure(data = data, layout = layout)
py.iplot(fig)
```

### Model Building

Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import roc_auc_score, roc_curve, scorer
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score, recall_score

# Split train and test set
train, test = train_test_split(df, test_size = .25, random_state = 123)

col = [x for x in df.columns if x not in y]
train_x = train[col]
train_y= train[y]
test_x = test[col]
test_y= test[y]
```

```python
logit = LogisticRegression()
logit_fit = logit.fit(train_x, train_y)
```

```python
pred = logit.predict(test_x)
prob = logit.predict_proba(test_x)
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(test_y, pred)

print(conf_matrix)
```

```python
# ROC Curve
roc_auc = roc_auc_score(test_y, pred)
print('Area under curve : ', roc_auc)
```

```python
# Accuracy score
accuracy_score(test_y, pred)
```

```python
# Classification Report
report = classification_report(test_y, pred)
print(report)
```
